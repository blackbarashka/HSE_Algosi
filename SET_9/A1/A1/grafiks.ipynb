# Полностью неупорядоченные наборы строк

```python
import pandas as pd
import matplotlib.pyplot as plt


csv_file = 'random_results.csv'
df = pd.read_csv(csv_file)



# Выделим размеры массива
sizes = df['ArraySize']

algorithms = [
    "QuickSort",
    "MergeSort",
    "STRINGQUICKSORT",
    "STRINGMERGESORT",
    "MSDRadix",
    "MSDRadix_Quick"
]

# Построим график времени для каждого алгоритма
plt.figure(figsize=(10, 6))
for alg in algorithms:
    plt.plot(sizes, df[f'{alg}_TimeMs'], marker='o', label=alg)

plt.title('Time Comparison (ms)')
plt.xlabel('Array size')
plt.ylabel('Average time (ms)')
plt.grid(True)
plt.legend()
plt.show()

# Построим график сравнений для каждого алгоритма
plt.figure(figsize=(10, 6))
for alg in algorithms:
    plt.plot(sizes, df[f'{alg}_Comparisons'], marker='o', label=alg)

plt.title('Character-by-character Comparisons')
plt.xlabel('Array size')
plt.ylabel('Average number of comparisons')
plt.grid(True)
plt.legend()
plt.show()
```
